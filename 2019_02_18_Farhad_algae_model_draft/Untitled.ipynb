```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
# #############################################################################
# read the data 
xlsx = pd.ExcelFile('Book2.xls')
xvalues = pd.read_excel(xlsx,0, usecols="A:H",header=None)
yvalue= pd.read_excel(xlsx,0, usecols="I",header=None)
X = xvalues.values
y = yvalue.values
#################################################
# Fit regression model
svr_rbf = SVR(kernel='rbf', C=100, gamma=.001)
y_rbf = svr_rbf.fit(X, y).predict(X)
# #############################################################################
plt.plot(y_rbf,'b')
#hold(True) 
plt.plot(y,'r')
```
