```python
import numpy as np
import matplotlib.pyplot as plt

wo = 15
l = 100
E = 1400
A = 25

w = np.array([np.linspace(-wo, wo, 1000)])
#print(w)

Pnd = w/l * ((wo/l)**2 - (w/l)**2)
P = E * A * Pnd
#print(Pnd)

plt.plot(w[0], Pnd[0])
plt.plot(w[0], P[0])

maxima = E * A * np.max(Pnd[0])
minima = E * A * np.min(Pnd[0])

Pmax = E * A * 2* (wo / l)**3 / (3*np.sqrt(3))




print(f"{np.round(maxima, 2)}  \n{np.round(minima, 2)} \n{np.round(Pmax, 2)}")
print(f"P = {np.round(Pmax, 2)} Newtons")
print(f"Displacement = {} Newtons")
```
