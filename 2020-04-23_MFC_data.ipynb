Define experiment time periods

```python
Exp1 = [65, 450]
Exp2 = [1000, 2100]
Exp3 = [2080, 2800]#[2260, 2800]
Exp4 = [3570, 4220]
```

```python
def plot_exp(Exp, MFC, timestep, substrate='Glucose_10mM'):
    exp = slice(Exp[0], Exp[1])
    plt.plot(MFC[exp])

    step = int(timestep * 60 / 2)
#     print(step)
    
#     print(list(range(Exp[0], Exp[1], step)))
#     print(list(range(Exp[0], Exp[1], step))[:-1:2])

    for t in list(range(Exp[0]+1, Exp[1], step))[:-1:2]:
        #data.Glucose_10mM[t : t+step] = 0
        data[substrate][t : t+step] = 0
        

    plt.plot(data[substrate][exp]*100)
```

```python
plot_exp(Exp1, data.MFCs, 2)   
```

```python
plot_exp(Exp2, data.MFCl, 5)  
```

```python
plot_exp(Exp3, data.MFCs, 3, substrate='water')  
```

```python
plot_exp(Exp4, data.MFCs, 3)  
```

```python
def nice_plot(Exp, MFC, title, substrate='Glucose_10mM'):
    
    exp = slice(Exp[0], Exp[1])

    fig, ax1 = plt.subplots()

    c = 'tab:red'
    ax1.set_xlabel('time (s)')
    ax1.set_ylabel(substrate, color=c)
    ax1.plot(data.t_s[exp], data[substrate][exp], color=c)
    ax1.tick_params(axis='y', labelcolor=c)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    c = 'tab:blue'
    ax2.set_ylabel('MFC voltage (V)', color=c)  # we already handled the x-label with ax1
    ax2.plot(data.t_s[exp], data[MFC][exp], color=c)
    ax2.tick_params(axis='y', labelcolor=c)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped

    ax2.set_title(title)
```
